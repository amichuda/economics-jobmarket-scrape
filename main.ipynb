```python
# Once we download the file, find the file `resultset` in the Downloads folder with newest date tag and load it
DOWNLOADS = "/Users/lordflaron/Downloads/"
downloads_folder = Path(DOWNLOADS)

list_of_files = downloads_folder.glob("joe_resultset*")
newest_joe_file = max(list_of_files, key=os.path.getctime)

```

```python
joe_listings = (
    pd.read_excel(newest_joe_file)
    .astype(str)
    .map(str.lower)
    .query("not JEL_Classifications.str.contains('macro')")
    .drop(columns = ['jp_agency_insertion_num'])
    .astype({'Application_deadline' : '<M8[ns]'})
    .query("Application_deadline > @pd.to_datetime('today')")
    .sort_values('Application_deadline')
    )
```
